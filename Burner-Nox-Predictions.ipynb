```python
import statsmodels.api as sm
```

```python

#Y = [1,3,4,5,2,3,4]
#X = range(1,8)
```

```python

burner_data_wna_na = burner_data_wna.dropna()

y = burner_data_wna_na["mov-average-Nox"]
X = burner_data_wna_na[["PA-Outlet Pressure","Kiln-IL Pressure","Kiln - Inlet Temp"]]
sm.add_constant(X)
```

```python
burner_data_wna_na
```

```python

est = sm.OLS(y,X)
```

```python
results = est.fit()
```

```python
results.summary()
```

```python
burner_data_wna_na["PredictedNox"]=results.predict(X)
```
