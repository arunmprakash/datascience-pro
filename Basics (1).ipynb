```python
import pandas as pd
```

```python
movies = pd.read_csv("fandango_score_comparison.csv")
```

```python
movies
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.hist(movies["Fandango_Stars"])
plt.show()
```

```python
plt.hist(movies["Metacritic_norm_round"])
```

```python
import numpy as np
fandango_mean = movies["Fandango_Stars"].mean()
Metacritic_mean = movies["Metacritic_norm_round"].mean()
fandango_median = movies["Fandango_Stars"].median()
Metacritic_median = movies["Metacritic_norm_round"].median()
fandango_sd = np.std(movies["Fandango_Stars"])
Metacritic_sd = np.std(movies["Metacritic_norm_round"])
print(fandango_mean,Metacritic_mean,fandango_median,Metacritic_median,fandango_sd,Metacritic_sd)
```

```python
plt.scatter(movies["Fandango_Stars"],movies["Metacritic_norm_round"])
```

```python
movies["fm_diff"] = abs(movies["Metacritic_norm_round"]-movies["Fandango_Stars"])
```

```python
movies.sort("fm_diff",ascending = False)
```

```python
print(movies.head(5))
```

```python
from scipy.stats import pearsonr
```

```python
pearsonr(movies["Fandango_Stars"],movies["Metacritic_norm_round"])
```

The fairly low correlation implies that the two sets are not linearly correlatd;. the positive numbers imply that the increase in Fandango rating proportionally increases metacritic ratings also.


```python
from scipy.stats import linregress
slope,intercept,r_value,p_value,std_err = linregress(movies["Fandango_Stars"],movies["Metacritic_norm_round"])
```

```python
print(slope,intercept,r_value,p_value,std_err)
```

```python
x = [3.0, 4.0]
```

```python
y = [slope*i+intercept for i in x]
```
